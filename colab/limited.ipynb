# **Пределы и непрерывность функций**

## **1. Функции**

**Функция** — это правило, сопоставляющее каждому элементу области определения **ровно одно** значение из области значений.

Проще говоря, функция принимает на вход некоторое значение и выдаёт соответствующее ему выходное значение.

Обозначается обычно $ f(x) $: $ x $ — то, что подаём на вход, $ f(x) $ — то, что получаем на выходе.
/
- **Область определения** — все допустимые входные значения.
- **Область значений** — все возможные выходные значения.

> Множество входных значений ещё называют областью определения, а множество выходных значений — областью значений.

**Пример:** ( $f(x) = 2x $).

Если ($ x = 3 $), то ($ f(3) = 6 $);

если ($ x = 5 $), то ($ f(5) = 10 $).

Каждому ($ x $) соответствует единственный результат.


### **Основные типы функций**

| Тип | Формула | Когда применяется |
|-----|---------|-------------------|
| Линейная | $ f(x) = mx + b $ | Прямая пропорциональность (цена от количества) |
| Квадратичная | $ f(x) = ax^2 + bx + c $ | Параболические зависимости (затраты от объёма) |
| Экспоненциальная | $ f(x) = a e^{kx} $ | Рост популяции, радиоактивный распад |
| Логарифмическая | $ f(x) = a \ln(bx) $ | Затухающий эффект (насыщение рекламы) |
| Гиперболическая | $ f(x) = a / x $ | Обратная пропорциональность (давление от расстояния) |

В машинном обучении функции используются для моделирования зависимостей, оценки ошибок и настройки параметров.


**Функции особенно важны в анализе данных и машинном обучении, поскольку они:**

- Позволяют моделировать зависимость между входными признаками и целевой переменной.
- Помогают оценить качество модели (через функцию потерь).
- Лежат в основе алгоритмов оптимизации, которые настраивают параметры моделей, чтобы добиться их наилучшего функционирования.

## **2. Предел и непрерывности функций**

### **2.1 Предел последовательности**

#### **Отличие функции от последовательности**

**Простыми словами**

**Последовательность** — это функция, у которой аргумент бывает только натуральным (дискретный). Обычная функция живёт на непрерывной прямой. Поэтому график последовательности — это отдельные точки, а график функции — сплошная линия (если функция непрерывна).

Часто алгоритмы машинного обучения работают **итеративно**: на каждом шаге параметры чуть меняются, приближаясь к оптимальным. Возникает последовательность значений ($a_1, a_2, a_3...$). Важно понять, **к чему она стремится** — это и есть предел последовательности.


#### **Определение**




> Это основа для разработки надёжных, устойчивых и эффективных алгоритмов.



> **Предел последовательности** — это значение, к которому приближаются элементы последовательности по мере роста их номеров.

Формально последовательность чисел ${a_n}$ имеет конечный предел A (записывают $ \lim_{n\to\infty} a_n = A $) если для любого сколь угодно малого положительного числа $ϵ >0$ существует такое натуральное число N, что для всех  $n > N$ выполняется неравенство:

$ |a_n - A| < ϵ $

> **Смысл определения предела:** для любой заданной точности ε мы можем указать такой номер N, что начиная с него все члены последовательности будут отличаться от предела меньше чем на ε. То есть мы не просто находим N для одного конкретного ε, а должны уметь делать это для каждого ε, каким бы маленьким оно ни было. Если это возможно — предел существует, и последовательность действительно «стремится» к этому числу.

#### **Пример: предел последовательности ($a_n = 1/n$)**


Возьмём последовательность: $( a_n = \frac{1}{n} $).  
С ростом $( n $) значения стремятся к 0. Докажем это формально.

Нужно для любого $ϵ > 0 $ найти такой номер N, что для всех $ n > N$ выполняется $|a_n - 0| < ϵ$.

Распишем неравенство:
$[
\left| \frac{1}{n} - 0 \right| < \varepsilon \quad \Leftrightarrow \quad \frac{1}{n} < \varepsilon.
$]
Так как $n > 0$, это равносильно $ n > \frac{1}{\varepsilon} $.

Итак, если мы возьмём n больше $\frac{1}{\varepsilon}$, то нужное неравенство будет выполнено. Однако номер N должен быть **натуральным числом**, а число $ \frac{1}{\varepsilon}$ не обязано быть целым. Поэтому выберем
$N= \left\lfloor \frac{1}{\varepsilon} \right\rfloor + 1,
$
где $\lfloor x \rfloor $ — целая часть числа  $x$ (округление вниз). Тогда $ N $ — натуральное и $ N > \frac{1}{\varepsilon}$.

Теперь для любого $ n > N $ имеем:
$
\left| \frac{1}{n} - 0 \right| = \frac{1}{n} < \frac{1}{N} < \varepsilon.
$
Таким образом, мы доказали, что $\lim_{n\to\infty} \frac{1}{n} = 0$.

#### **Объяснение примера**

Объяснение строится вокруг определения предела: мы должны для любого заданного $ε > 0$ найти такой номер N, начиная с которого все члены последовательности будут отличаться от предела меньше чем на ε. В данном случае предел A = 0, поэтому условие $|a_n - A| < ε$ превращается в $|1/n - 0| < ε$, то есть $1/n < ε$ (так как n положительно).

Теперь наша задача — понять, какие n удовлетворяют неравенству $1/n < ε$. Перевернём его (при положительных n и ε): $n > 1/ε$. Это значит, что если мы возьмём любое n, большее чем $1/ε$, то автоматически получим 1/n < ε.

Проблема в том, что номер n должен быть **натуральным числом** (1, 2, 3, ...), а число $1/ε$ не обязано быть целым. Поэтому мы не можем просто сказать N = 1/ε — вдруг это дробное? Нам нужно выбрать такое натуральное N, чтобы все n > N точно были больше 1/ε. Самый простой способ — взять целую часть от 1/ε (обозначается [1/ε] или floor(1/ε)) и прибавить 1. Это гарантирует, что N будет строго больше 1/ε (поскольку [x] ≤ x < [x]+1). Теперь для любого n > N имеем n ≥ N+1 > 1/ε, значит, n > 1/ε, и тогда 1/n < ε, что и требовалось.

Итак, выбор N = [1/ε] + 1 — это просто формальный способ сказать: «возьмём номер, который гарантированно больше, чем 1/ε». Тогда для всех последующих номеров неравенство выполняется. Это работает для любого ε, каким бы маленьким оно ни было: если ε = 0.0001, то 1/ε = 10000, N = 10001, и при n > 10000 уже 1/n < 0.0001. В этом и заключается смысл предела: члены последовательности могут быть сколь угодно близки к нулю, стоит только уйти достаточно далеко по номерам.

### **2.2 Предел функции в точке**

Предел функции описывает поведение её значений при приближении аргумента к определённой точке.

#### **Определение**

Если значения функции становятся всё ближе к конкретному числу $L$ по мере приближения аргумента $x$ к $x_0$, говорят, что функция имеет предел L в точке $x_0$.

**Строгое определение:**

Функция $f(x)$ имеет предел в точке $L$ (записывают $\lim_{x \to x_0} f(x) = L$), если для любого $\varepsilon > 0$ существует $ δ > 0$, такое что если $0 < |x - x_0| < δ $, выполняется неравенство:

 $ |f(x) - L| < \varepsilon $.  

То есть значения функции можно сделать сколь угодно близкими к $L$, выбирая $x$ достаточно близко к $x_0$, не не равен ему.

#### **Пример**

$\lim_{x \to 2} (3x - 1) = 5 $

Для любого $\varepsilon > 0$ положим $\delta = \varepsilon / 3 $.

Если $ 0 < |x - 2| < \delta $, то

$ |(3x-1) - 5| = 3|x-2| < 3\delta = \varepsilon $.

### **2.3 Односторонние пределы**

#### **Определение**

Когда функция приближается к точке с разных сторон, её поведение может быть различным. Для описания таких ситуаций существуют понятия **левостороннего и правостороннего** пределов.

- **Левосторонний предел** $\lim_{x \to x_0^-} f(x)$ — $x$ приближается к $x_0$ со стороны меньших значений ($x < x_0$).
- **Правосторонний предел** $\lim_{x \to x_0^+} f(x)$ — $x$ приближается справа ($x > x_0$).

#### **Пример 1: ReLU**

Возьмем функцию активации в нейронных сетях ReLU: $f(x) = \max(0, x)$

  
Сама функция непрерывна в 0: слева 0, справа 0. Но её производная (которая не является частью функции, а новой функцией) имеет разрыв: левый предел производной 0, правый — 1.

#### **Пример 2: Huber Loss**

Рассмотрим функцию ошибки Huber Loss, которая используется для задач регрессии. Она **сочетает в себе свойства квадратичной и линейной функций**, что **делает её устойчивой к выбросам**. Вот определение Huber Loss:

$
L_\gamma(a) =
\begin{cases}
\frac{1}{2} a^2, & |a| \le \gamma \\
\gamma (|a| - \frac{1}{2}\gamma), & |a| > \gamma
\end{cases}
$

$a = y - \hat{y} $ — разница между истинным значением $y$ и предсказанием модели $\hat{y}$.

$γ>0$ — параметр, определяющий область перехода между квадратичной и линейной частями функции (порог).

##### **Почему коэффициенты 1/2?**

Коэффициенты 1/2 нужны, чтобы наша функция не имела скачков, а также чтобы наша функция скорости изменения ошибки (производная первого порядка) также не имела скачков. **Это нужно чтобы наши функции были непрерывны**

image.png

##### **Анализ функции**

- При $|a| \le \gamma$ функция совпадает с квадратичной функцией потерь $\frac{1}{2}a^2$.  
  Это даёт высокую чувствительность к малым ошибкам и позволяет эффективно минимизировать небольшие отклонения.

- При $|a| > \gamma$ функция становится линейной (с наклоном $\gamma$).  
  Это снижает влияние крупных ошибок и выбросов, делая модель более устойчивой.

##### **Непрерывность**

В точках $a = \gamma$ и $a = -\gamma$ функция Huber Loss остаётся непрерывной и имеет непрерывную первую производную.

- Левосторонняя производная при $a \to \gamma^-$ равна $ \gamma $.  
- Правосторонняя производная при $a \to \gamma^+ $ равна $ \gamma $.

Таким образом, первая производная непрерывна в точках стыковки.  
Однако **вторая производная** испытывает скачок:

$
L_\gamma''(a) =
\begin{cases}
1, & |a| < \gamma,\\
0, & |a| > \gamma,
\end{cases}
$

при этом в точках $ a = \pm\gamma $ вторая производная не определена (разрыв). Это означает, что функция не является гладкой второго порядка в окрестности $ \pm\gamma $, несмотря на непрерывность и дифференцируемость первой производной.

##### **Преимущества**

Благодаря такому устройству Huber Loss сочетает преимущества MSE (чувствительность к малым ошибкам) и MAE (устойчивость к выбросам), что делает её популярным выбором в задачах регрессии при наличии шума и аномалий в данных.

## **3. Непрерывность и свойства непрерывных функций**

**Непрерывность обеспечивает стабильность и предсказуемость поведения моделей**, позволяя понять, как малейшие изменения входных данных отражаются на результате. Это особенно важно для создания надёжных и эффективных алгоритмов.

### **Практическое применение**

- **Стабильность модели** — это способность алгоритма давать предсказуемые результаты при небольших изменениях входных данных. Малые возмущения на входе не должны приводить к существенным отклонениям на выходе, что особенно важно в условиях работы с шумными или неидеальными данными.

- **Устойчивость модели** — это её способность сохранять точность и производительность даже при наличии шума, выбросов или изменений в данных. Устойчивая модель демонстрирует хорошую обобщающую способность на новых данных и не склонна к переобучению.

- **Предсказуемость модели (Интерпретируемость)** — это её способность выдавать результаты, которые интуитивно или математически согласуются с изменениями входных данных. Например, если модель прогнозирует температуру на основе времени суток, мы ожидаем, что небольшое изменение времени не приведёт к резким скачкам температуры, как если бы предсказания показывали +30°C на рассвете и –10°C через минуту. Предсказуемость повышает доверие к модели.

Эти понятия служат основой для понимания важности непрерывности функций в математическом анализе и их роли в машинном обучении. Вернемся теперь к понятию непрерывности.

### **Разница устойчивости и стабильности**

Представь, что у нас есть модель, которая предсказывает температуру на завтра по историческим данным. У неё есть два важных свойства — стабильность и устойчивость, и это не одно и то же.

**Стабильность** — это про то, насколько плавно меняется прогноз при крошечных, почти незаметных изменениях на входе.  
Пример: ты ввёл в модель температуру сегодня в 20.0 °C, и она выдала прогноз 22.0 °C. Если ты чуть-чуть изменишь вход — скажем, подашь 20.1 °C, — стабильная модель предскажет что-то около 22.05 °C, то есть результат изменится незначительно. График такой модели будет гладким, без резких скачков. Это и есть непрерывность: малая вариация данных не ломает предсказание. Если модель нестабильна, то при 20.1 °C она может выдать 35 °C — и это катастрофа, особенно если ты планируешь урожай.

**Устойчивость** же отвечает на вопрос: «А что будет, если в данные попадёт настоящая грязь — выброс, ошибка измерений?»  
Допустим, в обучающей выборке случайно оказался один безумный день, когда термометр показал +99 °C из-за сбоя. Неустойчивая модель (например, обычная линейная регрессия без регуляризации) сильно «поведётся» на этот выброс: она изменит свои коэффициенты так, чтобы учесть эту аномалию, и в итоге начнёт завышать прогнозы для всех нормальных дней. Устойчивая же модель (скажем, с Huber Loss или регуляризацией) проигнорирует этот выброс или слегка скорректируется, но сохранит разумную точность на основной массе данных. То есть устойчивость — это способность не подстраиваться под шум и сохранять обобщающую способность.

Итак, главное различие:  
- **Стабильность** — про реакцию на *малые* изменения входных данных (непрерывность, гладкость). «Чуть тронул вход — чуть изменился выход».  
- **Устойчивость** — про реакцию на *большие* помехи, выбросы, шум в данных (робастность). «Мусор на входе не портит всю модель».

В идеале нам нужны и стабильность, и устойчивость одновременно, чтобы модель была надёжной в реальных условиях.

### **3.1 Определение непрерывности**

Интуитивно функция называется непрерывной в точке, если небольшие изменения аргумента приводят к небольшим изменениям значения функции. Другими словами, **функция ведёт себя предсказуемо, без резких скачков, в каждой точке своей области определения.**

#### **Строгое определение**

Функция $ f(x) $ называется **непрерывной в точке** $ x_0 $, если выполняются два условия:

1. Значение $ f(x_0) $ определено (существует).
2. Существует предел $ \lim_{x \to x_0} f(x) $, и он равен значению функции в этой точке:  
$
\lim_{x \to x_0} f(x) = f(x_0).
$

Если хотя бы одно из этих условий нарушается, функция называется **разрывной** в точке $ x_0 $.

#### **Пример**

Рассмотрим пример функции потерь, которая естественным образом возникает в задачах бинарной классификации.


Пусть $ a = y - \hat{y} $, где $ y $ — истинное значение целевой переменной, а $ \hat{y} $ — предсказание модели.  
Определим функцию потерь  

$
L(a) =
\begin{cases}
1, & a \neq 0,\\
0, & a = 0.
\end{cases}
$

В задачах бинарной классификации это означает: если предсказание совпадает с истиной ($ a = 0 $), потери равны нулю, иначе — единице.  
Такая функция интуитивно понятна и точно отражает число ошибок, однако **она не подходит для оптимизации градиентными методами**.

**Почему градиентный спуск не работает:**

- В точках, где $ a \neq 0 $, функция постоянна (1), её производная равна нулю — градиент не даёт информации о направлении улучшения.
- В самой точке $ a = 0 $ функция имеет разрыв: предел слева и справа не равен значению в нуле (или, если смотреть на все $ a $, то в нуле она равна 0, а сразу рядом — 1). Производная в точке разрыва не определена.

Таким образом, градиентный спуск не может корректно обновлять параметры модели, потому что либо градиент нулевой почти всюду, либо не существует.

**Решение:** на практике разрывные функции потерь заменяют непрерывными аппроксимациями, которые сохраняют смысл задачи, но допускают вычисление градиента.  
Примеры:

- **Логистическая функция потерь (log loss):** $ L(a) = \ln(1 + e^{-a}) $ — гладкая, выпуклая, подходит для классификации.
- **Hinge loss:** $ L(a) = \max(0, 1 - a) $ — непрерывна, но имеет излом, также используется в SVM.
- **Huber Loss** (для регрессии) — сочетает квадратичную и линейную части, непрерывна и дифференцируема.

Благодаря таким заменителям задача оптимизации становится разрешимой, а поведение модели остаётся близким к желаемому.

### **3.2 Основные свойства непрерывных функций**

- **"Арифметика"**.
 Если функции $ f(x) $ и $ g(x) $ непрерывны в точке $ x_0 $, то их сумма, разность, произведение и частное (при $ g(x_0) \neq 0 $) также непрерывны в этой точке. Это свойство делает возможным построение сложных моделей из простых компонентов. **Пример:** В нейронных сетях непрерывность сохраняется на всех уровнях благодаря тому, что линейные преобразования и функции активации (например, ReLU) являются непрерывными.

- **Композиция**.
  Если функция $ f(u) $ непрерывна в точке $ u_0 $, а функция $ u(x) $ непрерывна в точке $ x_0 $ и $ u_0 = u(x_0) $, то составная функция $ f(u(x)) $ также непрерывна в точке $ x_0 $.

### **3.3 Важные теоремы о непрерывных функций**

- **Теорема Больцано — Коши (о промежуточном значении)**  
Если функция $ f(x) $ непрерывна на отрезке $ [a, b] $ и принимает на его концах значения разных знаков ($ f(a) \cdot f(b) < 0 $), то существует хотя бы одна точка $ c \in (a, b) $, в которой $ f(c) = 0 $.

Эта теорема лежит в основе численных методов поиска корней уравнений (например, метод деления пополам).

- **Теорема Вейерштрасса**  
Непрерывная на отрезке $ [a, b] $ функция достигает на нём своего максимума и минимума.

Это свойство важно в задачах оптимизации: оно гарантирует, что у непрерывной функции на замкнутом ограниченном множестве существует наилучшее решение.

### **3.4 Скорость изменения функции**

> Одна лишь непрерывность **не ограничивает скорость изменения** функции. Функция может быть непрерывной, но при этом резко меняться, что может вызвать сложности в задачах оптимизации и анализа.

Именно здесь на помощь приходит понятие **Липшицевой непрерывности**, которое вводит более строгий контроль за тем, насколько сильно значения функции могут изменяться в ответ на изменения входных данных.

## **4. Липшицева непрерывность**

> **Липшицева непрерывность** — более строгая форма непрерывности, которая ограничивает скорость изменения функции. Это свойство важно для анализа сходимости и устойчивости алгоритмов.

### **Определение**

Функция $ f(x) $ называется **непрерывной по Липшицу** на множестве $ D $, если существует константа $ L > 0 $, такая, что для любых $ x_1, x_2 \in D $ выполняется:

$
|f(x_1) - f(x_2)| \le L \, |x_1 - x_2|.
$

> **Почему это важно для нас:** Липшицева непрерывность градиента функции потерь позволяет ограничивать шаг обучения в градиентном спуске, обеспечивая стабильность и быструю сходимость алгоритма, то есть уменьшение числа итераций, необходимых для достижения заданного уровня точности.

В алгоритмах оптимизации градиентного спуска с адаптивным шагом мы будем использовать именно предположение о непрерывности по Липшицу — чтобы обеспечить корректное поведение.

#### **Пояснение**

**Непрерывность по Липшицу градиента функции потерь** — это именно то свойство, которое позволяет **теоретически обосновать выбор шага обучения** в градиентном спуске и гарантировать сходимость.  

Зная константу Липшица $L$ для градиента $\nabla L(\theta)$, можно выбрать шаг $\alpha$ из безопасного интервала $(0, 2/L)$, и алгоритм будет сходиться к минимуму. Без этого условия пришлось бы подбирать шаг «наугад», рискуя расходимостью.  

> **Сходимость** — это свойство последовательности или итеративного процесса приближаться к определённому конечному значению (пределу) при неограниченном увеличении номера шага.  
В контексте градиентного спуска: параметры $ \theta_k $ сходятся, если с каждой итерацией они становятся всё ближе к оптимальному значению $ \theta^* $, минимизирующему функцию потерь.

> **Расходимость** — противоположное поведение: последовательность не стремится ни к какому конечному пределу.  
Она может неограниченно возрастать (уходить в $ +\infty $ или $ -\infty $), осциллировать без затухания или хаотически скакать.  
В градиентном спуске расходимость означает, что параметры или значение функции потерь не уменьшаются, а наоборот, «взрываются» и удаляются от минимума. Обычно это происходит при слишком большом шаге обучения $ \alpha $.

**Пример:**  
- Сходящаяся последовательность: $ a_n = 1/n $ стремится к 0.  
- Расходящаяся последовательность: $ a_n = (-1)^n $ колеблется между -1 и 1 и не имеет предела.  
- Расходимость градиентного спуска: $ \theta_{k+1} = \theta_k - 3\,\theta_k = -2\theta_k $; при $ \theta_0 \neq 0 $ значения будут неограниченно расти по модулю.

### **Пример**

Это пример, где **градиент функции не является Липшицевым на всей области определения, но становится таким при ограничении интервала** — это позволит лучше понять, как непрерывность по Липшицу зависит от области анализа.

#### **1. Рассмотрим функцию $ f(x) = e^x $.**

Её градиент (производная): $ \nabla f(x) = e^x $.

Проверим липшицевость градиента:

- На всей числовой прямой $ \mathbb{R} $ градиент $ e^x $ **не ограничен** (уходит в $ +\infty $ при $ x \to +\infty $), поэтому **не является липшицевым глобально**.
- Однако если ограничиться отрезком $ [0, 1] $, то можно найти локальную константу Липшица.


#### **2. Нахождение константы L на отрезке $[0,1]$**  


> Для дважды дифференцируемой функции константа Липшица градиента равна максимуму модуля второй производной:  

$ f''(x) = e^x $.  
На отрезке $[0,1]$:  

$
\max_{x \in [0,1]} f''(x) = e^1 = e \approx 2.718.
$

Тогда для любых $ x_1, x_2 \in [0,1] $:

$
|\nabla f(x_1) - \nabla f(x_2)| = |e^{x_1} - e^{x_2}| \le e \cdot |x_1 - x_2|.
$

Значит, градиент **липшицев на $[0,1]$** с константой $ L = e $.

#### **3. Выбор шага обучения**  

Безопасный шаг для градиентного спуска при липшицевом градиенте $ (0, 2/L) $:  

$
\alpha = \frac{1}{L} = \frac{1}{e} \approx 0.368.
$

#### **4. Итеративное обновление**  


$
x_{k+1} = x_k - \alpha \nabla f(x_k) = x_k - \frac{1}{e} \cdot e^{x_k} = x_k - e^{x_k - 1}.
$

#### **5. Поведение алгоритма**  

- Функция $ f(x) = e^x $ выпукла, но **не имеет глобального минимума на $ \mathbb{R} $** (она стремится к 0 при $ x \to -\infty $ и неограниченно растёт при ($ x \to +\infty $).
- Если начальная точка выбрана на отрезке $[0,1]$ и шаг $ \alpha = 1/e $, то итерации будут монотонно убывать, сходясь к точке минимума **на этом отрезке** (к левому концу $ x = 0 $), если минимизировать на $[0,1]$.
- В отсутствие ограничений градиентный спуск будет неограниченно двигаться влево ( $x \to -\infty $), так как функция убывает в том направлении; предела последовательности $x_k$ в $ \mathbb{R} $ не будет (расходимость к $ -\infty $).

#### **6. Важный факт:**  
> Для практической сходимости градиентного спуска **не требуется**, чтобы градиент был липшицевым на всей области определения. Достаточно, чтобы он обладал этим свойством **локально**, в той компактной области, где предположительно находится искомое решение (или куда мы можем «зажать» итерации). Это позволяет выбирать безопасный шаг и гарантировать сходимость даже для функций, глобальное поведение которых не является липшицевым.

## **Выпуклая функция**

**Выпуклая функция** (точнее, выпуклая вниз, или просто convex) — это функция, у которой график лежит **не выше отрезка**, соединяющего любые две точки графика.

Формальное определение: функция $ f $, определённая на выпуклом множестве (например, интервале), называется выпуклой, если для любых $ x_1, x_2 $ из области определения и любого $ \lambda \in [0,1] $ выполняется неравенство

$
(\lambda x_1 + (1-\lambda) x_2) \le \lambda f(x_1) + (1-\lambda) f(x_2).
$

Геометрически это означает, что секущая (хорда) всегда проходит выше графика. Если неравенство строгое при $ 0<\lambda<1 $ и $ x_1 \neq x_2 $, то функция называется **строго выпуклой**.

**Свойства (для дифференцируемых функций)**  
- Если $ f $ дифференцируема, то выпуклость равносильна тому, что касательная в любой точке лежит **не выше** графика:  
  $ f(y) \ge f(x) + f'(x)(y-x) $ для всех $ x,y $.  
- Если $ f $ дважды дифференцируема, то условие выпуклости: $ f''(x) \ge 0 $ для всех $ x $.

**Примеры**  
- $ f(x) = x^2 $ — строго выпуклая ( $f''(x)=2 > 0 $).  
- $ f(x) = e^x $ — строго выпуклая ( $f''(x)=e^x > 0 $).  
- $ f(x) = |x| $ — выпуклая, но не дифференцируемая в нуле.  
- $ f(x) = \ln x $ — **вогнутая** (выпуклая вверх) на $ x>0 $ ($ f''(x)=-1/x^2 < 0 $).  

**Зачем это в машинном обучении?**  

> Если функция потерь выпукла, то у неё **нет локальных минимумов, отличных от глобального**. Это гарантирует, что градиентный спуск (при правильном шаге) будет сходиться к наилучшему решению, а не застрянет в «яме». Многие стандартные функции потерь — выпуклые:
- MSE (среднеквадратичная ошибка) — выпукла.
- Log Loss (логистическая функция потерь) — выпукла.
- Huber Loss — выпукла.

> Для невыпуклых функций (например, нейронных сетей) сходимость гарантируется только к локальному минимуму, что требует более аккуратных методов оптимизации и инициализации.

В предыдущем примере с $ f(x)=e^x $ функция выпукла на всей числовой прямой, поэтому градиентный спуск движется в сторону убывания (к $ -\infty $), но из-за отсутствия глобального минимума на $ \mathbb{R} $ он расходится к $ -\infty $, а на ограниченном отрезке $ [0,1] $ сходится к левой границе, где достигается минимум на этом множестве.

### **Пример проверки выпуклости функции**

Рассмотрим функцию $ f(x) = x^2 $.

#### **Способ 1 (через вторую производную)**  


Для дважды дифференцируемой функции условие выпуклости: $ f''(x) \ge 0 $ при всех $ x $.  
Находим:  
$ f'(x) = 2x $,  
$ f''(x) = 2 $.  
Так как $ 2 > 0 $ для любого $ x $, функция $ f(x) = x^2 $ **строго выпукла** на всей числовой прямой.


#### **Способ 2 (по определению)**  

Проверим неравенство:  
$ f(\lambda x_1 + (1-\lambda) x_2) \le \lambda f(x_1) + (1-\lambda) f(x_2) $
для любых $ x_1, x_2 $ и $ \lambda \in [0,1] $.  

Подставим $ f(x)=x^2 $:  
$ (\lambda x_1 + (1-\lambda)x_2)^2 \le \lambda x_1^2 + (1-\lambda) x_2^2 $.  

Раскроем левую часть:  
$ \lambda^2 x_1^2 + 2\lambda(1-\lambda)x_1x_2 + (1-\lambda)^2 x_2^2 $.  

Перенесём правую часть влево и упростим:  
$ \lambda^2 x_1^2 - \lambda x_1^2 + 2\lambda(1-\lambda)x_1x_2 + (1-\lambda)^2 x_2^2 - (1-\lambda) x_2^2 \le 0 $.  
Заметим, что $ \lambda^2 - \lambda = \lambda(\lambda-1) = -\lambda(1-\lambda) $,  
$ (1-\lambda)^2 - (1-\lambda) = (1-\lambda)((1-\lambda)-1) = -\lambda(1-\lambda) $.  

Тогда получим:  
$ -\lambda(1-\lambda)x_1^2 + 2\lambda(1-\lambda)x_1x_2 - \lambda(1-\lambda)x_2^2 \le 0 $,  
$ -\lambda(1-\lambda)(x_1^2 - 2x_1x_2 + x_2^2) \le 0 $,  
$ -\lambda(1-\lambda)(x_1 - x_2)^2 \le 0 $.  

Поскольку $\lambda(1-\lambda) \ge 0 $ для $ \lambda \in [0,1] $ и $ (x_1 - x_2)^2 \ge 0 $, всё выражение действительно $ \le 0 $. Неравенство выполняется, функция выпукла (даже строго, если $ x_1 \neq x_2 $ и $ 0<\lambda<1 $).

**Пример 2: $ f(x) = e^x $**  
Вторая производная: $ f''(x) = e^x > 0 $ при всех $ x $, следовательно, $ e^x $ строго выпукла на $ \mathbb{R} $.

**Пример невыпуклой функции: $ f(x) = x^3 $**  
$ f''(x) = 6x $. При $ x < 0 $ вторая производная отрицательна, значит, функция не является выпуклой на всей прямой (она вогнута слева от нуля и выпукла справа).

## **5. Анализ сходимости алгоритмов**

Рассмотрим **градиентный спуск** — основной метод оптимизации. Параметры модели $ \theta $ обновляются по правилу:
> $
\theta_{k+1} = \theta_k - \alpha \nabla L(\theta_k),
$

**Цель:** последовательность $ \{\theta_k\} $ должна сходиться к минимуму $ \theta^* $. Анализ сходимости использует понятия предела и непрерывности.

### **Простой пример (линейная регрессия с 1 наблюдением)**

#### **1. Входные данные и формулы**

Наши входные данные:
| x | y |
|:---:|:---:|
| 1 | 0 |



**Линейная регрессия (наш случай):**

$ y = θ_1 * x_1 + θ_0 $

**Общий вид:**

$ y = \sum θ_i * x_i $, где $x_0 = 1$

#### **2. Распишем функцию потерь**

**Функция потерь**

**Основная формула**

$ MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $

**Формула для простого подсчета (будем использовать её)**

$ MSE = \frac{1}{2n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $

подставим значения, $n=1, x=1, y=0$:

$ L = \frac{(0 - \theta_1)^2}{2}$

найдем градиент функции потерь:

$ \nabla L = \theta_1$

#### **3. Распишем градиентный спуск**

$θ_{k+1} = \theta_k - (\alpha * \theta_k) $

$θ_{k+1} = \theta_k (1 - \alpha)$ - мы получили рекурентную формулу для параметра

**Распишем для множества значений и получим последовательность:**

$θ_{k+1} = \theta_k (1 - \alpha)$

$θ_{k+2} = \theta_k (1 - \alpha)^2$

$θ_{k+3} = \theta_k (1 - \alpha)^3$

...

$θ_{k+n} = \theta_k (1 - \alpha)^n$

#### **4. Распишем через предел**

**Наша задача чтобы градиентный спуск сошелся к минимуму. Распишем последовательность через предел**

Наша зачача, чтобы вес сходился к самому лучшему весу $\theta*$, в котором будет точка минимума, а значит градиент веса будет равен нулю.

$
\lim_{k \to ∞} θ_k = θ*.
$

**Распишем предел градиентов:**

$
\lim_{n \to ∞} θ_{k+n}(α - 1)^n = 0.
$

**Распишем градиенты по определению пределов**

$|θ_{k+n} * (\alpha - 1)^n - 0| < ϵ|$

$|θ_{k+n}| |(\alpha - 1)|^n < ϵ|$

#### **5. Рассмотрим 3 случая**

$|(\alpha - 1)|^n$

##### **5.1 Первый случай ($|α-1| > 1$)**

$\alpha > 2$ & $ \alpha < 0$

тогда, наша функция расходится:

$
\lim_{n \to ∞} θ_{k+n}(α - 1)^n = ∞.
$

##### **5.2 Второй случай ($|α-1| < 1$) (нужный вариант)**

$\alpha < 2$ & $ \alpha > 0$

тогда, наша функция сходится к 0, что нам и нужно:

$
\lim_{n \to ∞} θ_{k+n}(α - 1)^n = 0.
$

##### **5.3 Третий случай ($|α-1|=0,1$)**

$ \alpha = 2$ & $ \alpha = 0$

тогда мы просто стоим на месте

#### **6. Вывод**

Тем самым мы нашли диапозон нашего шага, при помощи **теоремы Липшица о непрерывности**.

> Если $ \alpha $ вне этого интервала ($\alpha ∈ (0, 2)$), последовательность расходится. Так предел последовательности и непрерывность градиента определяют успех оптимизации.

Это условие связано с непрерывностью градиента функции потерь
$L(θ)$. Для текущего примера константа Липшица
$L=1$, что позволяет явно установить границы на
α. Если α выходит за пределы этого диапазона, то обновления параметра становятся слишком резкими и последовательность
${θ_k}$ начинает расходиться.

#### **7. Доп объяснение**

**Почему предел последовательности $\theta_k$ равен нулю?**

В рассмотренном примере линейной регрессии функция потерь имеет вид  
$L(\theta) = \frac{1}{2}\theta^2 $.  
Это квадратичная парабола с вершиной в точке $\theta = 0$, которая и является её **глобальным минимумом**. Градиентный спуск стремится найти именно эту точку, так как в ней ошибка минимальна.

С аналитической стороны, итерационная формула дала  
$\theta_k = (1-\alpha)^k \theta_0 $.  
При выполнении условия сходимости $|1-\alpha| < 1 $ множитель $(1-\alpha)^k$ стремится к нулю с ростом \( k \). Тогда  
$
\lim_{k\to\infty} \theta_k = \theta_0 \cdot 0 = 0,
$
независимо от начального значения $\theta_0$. Таким образом, последовательность параметров сходится именно к точке минимума функции потерь.

Если бы функция потерь имела минимум в другой точке, скажем, $ \theta^* \neq 0 $, то и предел последовательности равнялся бы $\theta^* $. В данном простом случае минимум находится в нуле, поэтому предел — ноль.

### **5.1 Вывод**

Анализ сходимости показывает, как алгоритмы оптимизации приближаются к желаемому результату, шаг за шагом минимизируя ошибку. При этом ключевую роль играют пределы и непрерывность: они не только гарантируют, что процесс сходится, но и делают его предсказуемым и стабильным. Однако сходимость — это лишь часть задачи. Чтобы алгоритм был действительно эффективным, важно, чтобы модель оставалась устойчивой, даже если данные содержат шум или небольшие отклонения. Поговорим теперь об устойчивости моделей.

## **6. Устойчивость моделей**

> **Устойчивая модель** — это модель, результаты которой не меняются кардинально при небольших изменениях входных данных. Например, если мы добавим небольшой шум к нашим данным, устойчивая модель должна выдавать предсказания, которые остаются близкими к исходным.

Другое определение:

> **Устойчивость** — это свойство модели не менять резко свои предсказания при незначительных изменениях входных данных (например, при добавлении шума). Непрерывность модели гарантирует такую устойчивость, а липшицевость даёт количественную оценку: изменение выхода ограничено константой, умноженной на изменение входа.

**Как раз здесь мы и можем оценить роль непрерывности:**

- Непрерывность функции модели гарантирует, что малые изменения входных данных приводят к небольшим изменениям в предсказаниях модели. Это делает модель предсказуемой и надёжной.
- Если модель обладает непрерывностью по Липшицу, можно количественно оценить, насколько сильно изменится результат при небольших изменениях на входе.

#### **Пример (линейная регрессия с регуляризацией (L2))**

Функция потерь **включает штраф за большие веса**:

$
L(\theta) = \frac{1}{2n} \sum_{i=1}^n (y_i - \theta^T x_i)^2 + \frac{\lambda}{2} \|\theta\|^2.
$

Регуляризационный член $\frac{\lambda}{2} \|\theta\|^2 $ «сглаживает» решение, не давая параметрам становиться слишком большими. Это делает модель менее чувствительной к отдельным наблюдениям и шуму, то есть более устойчивой. Математически такая модель оказывается липшицевой с константой, зависящей от $ \lambda $.

> Регуляризация делает модель устойчивее, так как большие значения параметров подавляются, что предотвращает резкие изменения предсказаний.



## **7. Сводка ключевых определений**

- **Функция** $ f(x) $ — правило, сопоставляющее каждому элементу области определения ровно одно значение.
- **Область определения** — множество допустимых входных значений.
- **Область значений** — множество всех выходных значений.
- **Предел последовательности** $ \lim_{n\to\infty} a_n = A $: для любого $ \varepsilon > 0 $ существует $ N $, что $ n > N $ влечёт $ |a_n - A| < \varepsilon $.
- **Предел функции в точке** $ \lim_{x \to x_0} f(x) = L $: для любого $ \varepsilon > 0 $ существует $ \delta > 0 $, что из $ 0 < |x - x_0| < \delta $ следует $ |f(x) - L| < \varepsilon $.
- **Односторонние пределы** — левосторонний ($ x \to x_0^- $) и правосторонний ($ x \to x_0^+ $).
- **Непрерывность в точке:** $ \lim_{x \to x_0} f(x) = f(x_0) $.
- **Теорема Больцано – Коши:** если непрерывная на $ [a,b] $ функция принимает на концах значения разных знаков, то внутри отрезка есть корень.
- **Теорема Вейерштрасса:** непрерывная на отрезке функция достигает минимума и максимума.
- **Липшицева непрерывность:** существует $ L > 0 $ такая, что для всех $ x_1, x_2 $ из области определения $ |f(x_1) - f(x_2)| \le L |x_1 - x_2| $. Ограничивает скорость изменения функции.
- **Константа Липшица** $ L $ — наименьшее число, удовлетворяющее условию выше.
- **Градиентный спуск** — итеративный метод: $ \theta_{k+1} = \theta_k - \alpha \nabla L(\theta_k) $.
- **Устойчивость модели** — малое изменение входных данных не приводит к резкому изменению предсказания; часто обеспечивается непрерывностью или липшицевостью модели.
- **Регуляризация** — добавление штрафа к функции потерь для предотвращения переобучения и повышения устойчивости (например, $ \lambda \|\theta\|^2 $).
